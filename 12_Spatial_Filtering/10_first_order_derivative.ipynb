```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a image with black background and white circle
img = np.zeros((500,500),dtype='uint8') 
cv2.circle(img,(250,250), 150, (255,255,255), -1)
 
# Output dtype = cv2.CV_8U
sob_8u = cv2.Sobel(img,cv2.CV_8U,1,0,ksize=3)
 
# Output dtype = cv2.CV_64F. Then take its absolute and convert to cv2.CV_8U
sobel_64f_1 = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
abs_64f_1 = np.absolute(sobel_64f_1)
sobel_8u_1 = np.uint8(abs_64f_1)

# Output dtype = cv2.CV_64F with normalization.
sobel_64f_2 = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
abs_64f_2 = np.absolute(sobel_64f_2)
sobelx_8u1_2 = abs_64f_2/abs_64f_2.max()*255
sobel_8u_2 = np.uint8(sobelx_8u1_2)

plt.subplot(2,2,1)
plt.title("input image")
plt.imshow(img, cmap ='gray')

plt.subplot(2,2,2)
plt.title("sob_8u")
plt.imshow(sob_8u, cmap ='gray')

plt.subplot(2,2,3)
plt.title("sobel_8u_1(64F) without normalization")
plt.imshow(sobel_8u_1, cmap ='gray')

plt.subplot(2,2,4)
plt.title("sobel_8u_2(64F) with normalization")
plt.imshow(sobel_8u_2, cmap ='gray')

plt.gcf().set_size_inches(12, 12)
plt.show()
```
